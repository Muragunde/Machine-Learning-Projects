## Stock Sentiment Analysis using News Headlines

#### Importing Necessary Libraries

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

#### Reading the Dataset

```python
# Reading the Dataset
df = pd.read_csv(r'C:\users\hp\downloads\Data.csv',encoding='ISO-8859-1')
df.head()
```

#### Spliting the Data

```python
# Split the data into train and test data
train = df[df['Date']<'20150101']
test = df[df['Date']>'20141231']
```

```python
test.head()
```

#### Cleaning and Preprocessing the data

```python
# Preprocessing the dataset
# Replacing the values if it doesn't object
data = train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ", regex=True, inplace=True)

# Change the column names and assign index number
list1 = [i for i in range(25)]
new_index = [str(i) for i in list1]
data.columns = new_index
data.head()
```

```python
# Lowering the text data
for index in new_index:
    data[index] = data[index].str.lower()

data.head()
```

```python
' '.join(str(x) for x in data.iloc[1,0:25])
```

```python
headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[1]
```

#### Applying Bagofwords and model training with RandomForestClassifier

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
cv = CountVectorizer(ngram_range=(1,1))
traindataset=cv.fit_transform(headlines)
```

```python
rf = RandomForestClassifier(n_estimators=200, criterion='entropy')
rf.fit(traindataset, train['Label'])
```

```python
test_transform = []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,0:25]))
testdataset = cv.transform(test_transform)
predictions = rf.predict(testdataset)
```

#### Model Evaluation using Classificaion_report, confusion_matrix and Accuracy_score

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
