#### Importing Necessary Libraries

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report, accuracy_score
```

#### Read the Dataset Twitter Sentiment140 dataset

```python
df = pd.read_csv(r'C:\users\hp\downloads\training.1600000.processed.noemoticon.csv', encoding='latin-1', header=None)
## Specific Columns I picked
df = df[[0, 5]]
## renaming the column names
df.columns = ['polarity', 'text']
print(df.head())
```

```python
df.head()
```

#### Keep Only Positive and Negative Sentiments

```python
df = df[df['polarity'] != 2]

df['polarity'] = df['polarity'].astype(int)
df['polarity'] = df['polarity'].map({0: 0, 4: 1})

print(df['polarity'].value_counts())
```

```python
df.columns
```

#### Clean the Tweets

```python
def clean_text(text):
    return text.lower()

df['clean_text'] = df['text'].apply(clean_text)

print(df['clean_text'].head())
```

#### Split the Dataset into Training and Testing Data

```python
X_train, X_test, y_train, y_test =train_test_split(df['clean_text'], df['polarity'], test_size=0.2, random_state=42)
print(X_train.head())
```

```python
X_test.head()
```

#### Applying the TFIDF Vectorizer using scikit-learn Library

```python
vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1,2))

X_train_tfidf  = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

print('X_train TFIDF is: ', X_train_tfidf)
print('X_test TFIDF is: ',X_test_tfidf)
```

#### Model_Training and Prediction Using BernoulliNB Algorithm

```python
bnb = BernoulliNB()
bnb.fit(X_train_tfidf, y_train)

bnb_pred = bnb.predict(X_test_tfidf)

print("Bernoulli Naive Bayes Accuracy:", accuracy_score(y_test, bnb_pred))
print("\nBernoulliNB Classification Report:\n", classification_report(y_test, bnb_pred))
```

#### Train Support Vector Machine (SVM) model

```python
svm = LinearSVC(max_iter=1000)
svm.fit(X_train_tfidf,y_train)

svm_pred = svm.predict(X_test_tfidf)


print("Linear Support Vector Machine Accuracy:", accuracy_score(y_test, svm_pred))
print("\nLinear Support Vector Machine Classification Report:\n", classification_report(y_test, svm_pred))
```

```python
#### Train Logistic Regression model
```

```python
logreg = LogisticRegression(max_iter=100)
logreg.fit(X_train_tfidf,y_train)

logreg_pred = logreg.predict(X_test_tfidf)

print("Logistic Regression Accuracy:", accuracy_score(y_test, logreg_pred))
print("\nLogistic Regression Classification Report:\n", classification_report(y_test, logreg_pred))
```

#### Making a Prediction on Some Samples

```python
sample_tweets = ["I love this!", "I hate that!", "It was okay, not great."]
sample_vec = vectorizer.transform(sample_tweets)

print("\nSample Predictions:")
print("BernoulliNB:", bnb.predict(sample_vec))
print("SVM:", svm.predict(sample_vec))
print("Logistic Regression:", logreg.predict(sample_vec))
```
