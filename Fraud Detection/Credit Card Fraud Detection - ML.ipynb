#### Plotting Correlation Map

```python
corrmat = df.corr()
plt.figure(figsize=(12,9))
sns.heatmap(corrmat, vmin=.8, square=True)
plt.show()
```

#### Preparing Data like Independent and Dependent Features

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
print(X.shape)
print(y.shape)

X_data = X.values
y_data = y.values
```

#### Split the Data into Training and Testing Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(X_data, y_data, test_size=0.2, random_state=42)
```

#### Model Training and Predicting the Class

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)

y_pred = rfc.predict(X_test)
```

#### Evaluating the Model using Metrics and using Heatmap create a confusion_matrix

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, f1_score, matthews_corrcoef
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
mcc = matthews_corrcoef(y_test, y_pred)

print("Model Evaluation Metrics:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print(f"Matthews Correlation Coefficient: {mcc:.4f}")

conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",
            xticklabels=['Normal', 'Fraud'], yticklabels=['Normal', 'Fraud'])
plt.title("Confusion Matrix")
plt.xlabel("Predicted Class")
plt.ylabel("True Class")
plt.show()
```
